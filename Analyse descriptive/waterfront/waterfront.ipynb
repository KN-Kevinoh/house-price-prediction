```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from collections import Counter, defaultdict
import math
```

```python
data = pd.read_csv("kc_house_data.csv")
data.head(10)
```

```python

floors = Counter(data['waterfront'])
print(floors)

keys = {0:"Non", 1:"Oui"}

plt.figure(figsize = (8, 8))
x = floors.values()

plt.pie(x, labels = [keys[x] for x in floors.keys()], 
        autopct = lambda x: str(round(x, 2)) + '%', 
        shadow = True)
plt.title("Maison au bord de la mer")
plt.legend()
plt.savefig('waterfront.png', transparent = True)
plt.show()
plt.close()
```
